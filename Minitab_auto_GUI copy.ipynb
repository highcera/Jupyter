```python
# 미니탭 반복 입력, 반복수 입력 필수

import pyautogui as ag
import time

# pyautogui.mouseInfo()

noData = input('# of combination :' ) 

time.sleep(3)
for i in range(2, int(noData)+1):
    ag.moveTo(230,62)
    ag.click()
    time.sleep(1)

    ag.moveTo(995,445)
    ag.click()
    ag.press('backspace', presses=8)
    ag.typewrite('Time'+str(i))
    time.sleep(1)

    ag.moveTo(985,552)
    ag.click()
    ag.press('backspace', presses=8)
    ag.typewrite('Temp'+str(i))
    time.sleep(1)

    ag.moveTo(985,590)
    ag.click()
    ag.press('backspace', presses=8)
    ag.typewrite('Vot'+str(i))
    time.sleep(1)

    ag.moveTo(1185,400)
    ag.click()
    time.sleep(1)
    ag.moveTo(977,500)
    ag.click()
    ag.press('backspace', presses=8)
    ag.typewrite('Censor'+str(i))
    ag.moveTo(1040,617)
    # time.sleep(1)
    ag.click()
    time.sleep(1)

    ag.moveTo(1190,665)
    ag.click()
    time.sleep(1)
```
