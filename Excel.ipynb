```python
import pandas as pd
sales = pd.read_excel('SupermarketSales.xlsx')
# 새로운 엑셀 파일을 불러옵니다.
# PrdManager.xlsx 파일을 읽어와 그 결과를 prd_manager라는 이름의 변수에 저장합니다.
prd_manager = pd.read_excel('PrdManager.xlsx')
# 결과를 출력합니다.
prd_manager
```

```python
sales = pd.merge(sales, prd_manager, how='left', on='제품군')
sales
```

```python
# merge() 함수는 두 데이터프레임의 공통 컬럼 또는 인덱스를 기준으로 두 개의 데이터를 합침. 기준 컬럼 : 키(key)

# Inner Join : 두 데이터프레임에서 키 값이 일치하는 데이터만 합침. how = 'inner'
# Outer Join : 두 데이터프레임의 모든 데이터를 합침. how = 'outer' 
# Left Outer Join : 데이터프레임의 A를 기준으로 A의 키 값을 충족하는 B의 데이터만 합침. how = 'left' 
# Right Outer Join : 데이터프레임의 B를 기준으로 B의 키 값을 충족하는 A의 데이터만 합침. how = 'right'
```

```python
# 엑셀 화일 각 시트 내 string 치환

import openpyxl as op

#path는 대상이 되는 엑셀파일의 절대 경로
def corretion(path : str):
    #openpyxl 워크북 정의
    wb = op.load_workbook(path)
    
    ws_list = wb.sheetnames  #해당 Workbook의 시트 목록을 리스트로 저장

    for  sht  in  ws_list:
        ws = wb[sht] #Sheet 객체 생성
    
        #해당 시트의 마지막 열, 마지막 행 
        column_max = ws.max_column
        row_max = ws.max_row
        #print (column_max, row_max)
     
        #열마다 행을 for loop문 진행
        for col_num in range(1, column_max+1):
            for row_num in range(1, row_max+1):
          
                #tempstr : cell값이 문자열이 아닌 경우를 감안하여 str로 바꿔줌
                tempstr = str(ws.cell(row = row_num, column = col_num).value)
                #문자열 함수 replace 사용(,를 빈칸으로)
                data = tempstr.replace(",","")
                #빈 셀의 경우 None이라는 문자열 타입이므로 제외하고 데이터 입력
                if data != "None":
                    ws.cell(row = row_num, column = col_num).value = data 
    
    #저장
    wb.save("P:\자동데이타생성\correction_result.xlsx")

if __name__ == "__main__":
    path = r"P:\자동데이타생성\before_correction.xlsx"

    #함수 실행
    corretion(path)
```
